```python
def max_product(s):
    """ 
    >>> max_product([10,3,1,9,2]) # 10 * 9
    90
    >>> max_product([5,10,5,10,5]) # 5 * 5 * 5
    125
    >>> max_product([])
    1
    """

    lst = [s]
    for e in lst:
        f = e * lst[e+2]
        return f



print(max_product(5, 3, 2))
```

```python
def remove_odd_indices(lst, odd, weight):
    """
    >>> s = [1, 2, 3, 4]
    >>> t = remove_odd_indices(s, True, 1)
    >>> s
    [1, 2, 3, 4]
    >>> t
    [1, 3]
    >>> l = [5, 6, 7, 8]
    >>> m = remove_odd_indices(l, False, 4)
    >>> m
    [24, 32]
    """


    lst2 = []


    if odd == True:
        for e in lst:
            if e % 2 == 0:
                e = ""
            else:
                lst2.append(e * weight)
        return lst2

    elif odd == False:
        for e in lst:
            if e % 2 != 0:
                e = ""
            else: 
                lst2.append(e * weight)
        return lst2
                


import doctest
doctest.testmod()
```

```python
def alternatingSum(lst):
    """
    >>> alternatingSum([5, 3, 8, 4])
    6
    >>> alternatingSum([])
    0
    """
    s = 0
    sum = 0
    for elem in lst:
        if s % 2 == 0:
            sum += elem
            s += 1

        else:
            sum -= elem
            s += 1

        

    return sum
import doctest
doctest.testmod()
```

```python
def two_of_three(i, j, k):
    """Return m*m + n*n, where m and n are the two smallest members of the
    positive numbers i, j, and k.
    >>> two_of_three(1, 2, 3)
    5
    >>> two_of_three(5, 3, 1)
    10
    >>> two_of_three(10, 2, 8)
    68
    >>> two_of_three(5, 5, 5)
    50
    """
    n = 0
    m = 0
    if i <= j or i<=k:
        n = i
    elif j<= i or j <= k:
        m = j
    elif k < i:
        n = i
    elif k < j:
        m = k
    x = (n*n) + (m*m)
    return x
import doctest
doctest.testmod()
```

```python
def largest_factor(n):
    """Return the largest factor of n (n > 1 !) that is smaller than n.
    A factor divides a number evenly.
    >>> largest_factor(15) # factors are 1, 3, 5
    5
    >>> largest_factor(80) # factors are 1, 2, 4, 5, 8, 10, 16, 20, 40
    40
    >>> largest_factor(13) # factor is 1 since 13 is prime
    1
    """
    m = 1
    while m  < n:
        if n % m == 0:
            x = m
            m+=1
        else:
            m +=1
    return x 
import doctest
doctest.testmod()
```
