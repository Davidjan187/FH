```python
def hasConsecutiveDigits(n):
    # Convert negative number to positive
    n = abs(n)
    
    # Base case: If n is a single-digit number, return False
    if n < 10:
        return False

    # Recursive case: Check if the last two digits are the same
    last_digit = n % 10
    second_last_digit = (n // 10) % 10

    if last_digit == second_last_digit:
        return True
    else:
        # Make a recursive call with the last digit removed
        return hasConsecutiveDigits(n // 10)

# Test the function
result = hasConsecutiveDigits(122345)
print(result)  # Output: False

result = hasConsecutiveDigits(122345)
print(result)  # Output: True
```

```python
def alternatingSum(L, counter=0, is_odd=True):
    if counter >= len(L):
        return 0

    current_num = L[counter]

    if is_odd and current_num % 2 != 0:
        return current_num + alternatingSum(L, counter + 1, not is_odd)
    elif not is_odd and current_num % 2 == 0:
        return current_num + alternatingSum(L, counter + 1, not is_odd)
    else:
        return alternatingSum(L, counter + 1, is_odd)

# Test the function
result = alternatingSum([4, 2, 5, 1, 2])
print(result)
```
